### Section 1.1: Making my Data Usable

In this section, I am reading in four separate files that contains data aggregated from each company. The data in each of these files seems to be formatted the same, so I developed a function that I can use to make my data easier to work with. This includes importing the data, turning the date strings into actual dates, and adding the company name to respective data so that I won't be confused as to which is which when I combine all the data.

```python
def mergable(fileName):
    df = pd.read_csv(fileName)
    df.Date = pd.to_datetime(df.Date, format='%d/%m/%Y')
    df.insert(loc = 0, column='Company', value=fileName.replace(".csv",""))
    return df
```

```python
hertz_data = mergable('50Hertz.csv')
```

```python
amprion_data = mergable('Amprion.csv')
```

```python
tenne_data = mergable('TenneTTSO.csv')
```

```python
transnet_data = mergable('TransnetBW.csv')
```

Now that I have all my data in the same format, I can merge it into one file.

```python
merged_data = pd.concat([hertz_data, amprion_data, tenne_data, transnet_data], ignore_index = True)
```

### Section 1.4: Analyzing the Day

After much success with aggregating previous data, our client now wants to be able to generate an area graph that depicts the TW/h Generated throughout a specified day for a given company. The client wants to be able to pass in the data to use, the company, the date, and they also want to be able to customize the color.

```python
def daily_view(dataframe, company, date, color):
    sub = dataframe[(dataframe['Company'] == company) & (dataframe['Date']==date)]
    sub.drop(['Company', 'Date'], axis=1,inplace=True)
    subCopy = sub.T
    sub1 = subCopy.plot.area(title ='Power Production: '+company+'/'+date, ylabel ='TW/h Generated', xlabel ='Time of Day', color=color, legend=False)
    return sub1
```

Since we are using our large dataframe for this, I know that both the company and the date have their own columns. Thus, I should be able to pick out the columns that match both of the criteria that my client passes in, and place it into its own dataframe.

For this section, the company and date will always be the same, so there isn't really a need for those columns. After dropping these, all I have for the columns are my times. When plotting these times, I noticed that my area chart wants to plot the index as the independent variable, but I want the independent variable to be the time. So, I can just transpose the data and switch which value is the column and which value is the row. I should then be able to make my graph.

```python
daily_view(merged_data, '50Hertz', '2019-8-24', 'purple')
```

#### Section 2.1.2: Visualizing this Information
After looking at this grouped data, I might be able to create a graph based off the carat and price of a diamond - but taking it a step further, is there a price difference between the cut of diamond? Generally, you might think that the better the cut of diamond, the more expensive the diamond might be.

I'm going to create subsets of this data by the cut of diamond, and see if the price depends on the carat of the diamond.

```python
ideal_cut = diamonds.query("cut == 'Ideal'")
ideal_cut.drop(['cut', 'color', 'depth', 'x', 'y', 'z'], axis=1, inplace=True)
ideal_cut = ideal_cut.groupby([ideal_cut['carat']]).mean()
```

```python
premium_cut = diamonds.query("cut == 'Premium'")
premium_cut.drop(['cut', 'color', 'depth', 'x', 'y', 'z'], axis=1, inplace=True)
premium_cut = premium_cut.groupby([premium_cut['carat']]).mean()
```

```python
vgood_cut = diamonds.query("cut == 'Very Good'")
vgood_cut.drop(['cut', 'color', 'depth', 'x', 'y', 'z'], axis=1, inplace=True)
vgood_cut = vgood_cut.groupby([vgood_cut['carat']]).mean()
```

```python
good_cut = diamonds.query("cut == 'Good'")
good_cut.drop(['cut', 'color', 'depth', 'x', 'y', 'z'], axis=1, inplace=True)
good_cut = good_cut.groupby([good_cut['carat']]).mean()
```

```python
fair_cut = diamonds.query("cut == 'Fair'")
fair_cut.drop(['cut', 'color', 'depth', 'x', 'y', 'z'], axis=1, inplace=True)
fair_cut = fair_cut.groupby([fair_cut['carat']]).mean()
```

```python
diamond_fig = plt.figure(figsize=(15,30))
ax5 = diamond_fig.add_subplot(515)
plt.plot(fair_cut.price, color ='red')
plt.title('Fair Diamond/Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
ax1 = diamond_fig.add_subplot(511, sharex=ax5, sharey=ax5)
plt.plot(ideal_cut.price, color='green')
plt.title('Ideal Diamond/Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
ax2 = diamond_fig.add_subplot(512, sharex=ax5, sharey=ax5)
plt.plot(premium_cut.price, color='yellow')
plt.title('Premium Diamond/Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
ax3 = diamond_fig.add_subplot(513, sharex=ax5, sharey=ax5)
plt.plot(vgood_cut.price, color='pink')
plt.title('Very Good Diamond/Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
ax4 = diamond_fig.add_subplot(514, sharex=ax5, sharey=ax5)
plt.plot(good_cut.price, color='blue')
plt.title('Good Diamond/Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price')
```
